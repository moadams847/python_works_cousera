```python

# webscraping exercise using beautiful soup and urllib

import sys
sys.path.append('/users/adams/anaconda3/lib/site-packages')

import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# ignore ssl certificate
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter- ')

if len(url) <= 0:
    url = 'http://www.dr-chuck.com'

# read the entire page
html = urllib.request.urlopen(url, context = ctx).read()
soup = BeautifulSoup(html, 'html.parser')
# print(soup)

# retrieve all of the ancho tags
# count = 0
tags = soup('a')
# print(tags)

for tag in tags:
    count = count + 1
    print(tag.get('href',None))
print(count, 'Iteration(s)')
```

```python
# web scraping exercise using url lib and beautiful
# scraping html data

import sys
sys.path.append('/users/adams/anaconda3/lib/site-packages')

import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import ssl

# ignore ssl certificate
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input('Enter- ')

if len(url) <= 0:
    url = 'http://py4e-data.dr-chuck.net/comments_42.html'

# read the entire page
html = urllib.request.urlopen(url, context = ctx).read()
soup = BeautifulSoup(html, 'html.parser')
# print(soup)

# # retrieve all of the ancho tags
# count = 0
tags = soup('span')

comments = list()
# print(tags)
for tag in tags:
    count = count + 1
    tag.get('class',None
#     print(tag.get('class',None))
    comment = tag.contents[0]
    comments.append(int(comment))
    
print(comments)
print(count, 'Iteration(s)')
```
